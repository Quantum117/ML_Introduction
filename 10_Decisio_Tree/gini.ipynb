```python
import numpy as np

def gini_impurity(y):
    """Calculate Gini impurity for a node"""
    if len(y) == 0:
        return 0
    proportions = np.bincount(y) / len(y)
    return 1 - np.sum(proportions ** 2)

# Example: Binary classification
pure_node = np.array([0, 0, 0, 0])  # All same class
mixed_node = np.array([0, 1, 2,4])  # Two classes 50/50

print(gini_impurity(pure_node))  # Output: 0.0 (perfect purity)
print(gini_impurity(mixed_node)) # Output: 0.5 (maximum impurity for binary)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x range from 0 to 1
x = np.linspace(0, 1, 100)  # 100 points between 0 and 1

# Calculate y = x + xÂ²
y = x - x**2

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, 'b-', linewidth=2)  # Blue solid line
plt.title('Plot of $f(x) = x + x^2$ from $x=0$ to $x=1$', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.grid(True, alpha=0.3)

# Show the plot
plt.show()
```
